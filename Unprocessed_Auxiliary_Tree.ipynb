```python
def subtree(li,branches):
    
    li_copy = li.copy()
    prevbranch = {}
    
    flag = 0
    
    #Construct branches------------------
                
    n=0
    for i in li_copy:
        flag = 0
        for j in range(n,len(li_copy)):
            if i == li_copy[j] and i == max(li_copy):
                flag = 0
            elif i == li_copy[j] and li_copy.index(i) == len(li_copy)-1:
                flag = 0
            elif i == li_copy[j]:
                flag = 0
                continue
            elif i<li_copy[j]:
                newbranch = {li_copy[j]:i}
                prevbranch = newbranch.copy()
                branches.append(prevbranch)
                n+=1
                flag = 1
            elif i>li_copy[j]:
                flag = 0
                continue
                
            if flag == 1:
                break
            elif flag == 0:
                newbranch = {0:i}
                prevbranch = newbranch.copy()
                branches.append(prevbranch)
                n+=1
                if li_copy.index(i) == len(li_copy)-1:
                    return [],branches
                else:
                    return li_copy[li_copy.index(i)+1:],branches
        
```

```python
def join_branches(dic,li):
    dic_copy = dic.copy()
    li_copy = li.copy()
    
    #Check if value == key---------------
    
    for k in dic_copy.keys():
        val = dic_copy[k]
        for n in range(0,len(val)):
            for i in li_copy:
                for l in i.keys():
                    if val[n] == l:
                        val[n] = i
    dic_copy[k] = val
    return dic_copy[k]
```
